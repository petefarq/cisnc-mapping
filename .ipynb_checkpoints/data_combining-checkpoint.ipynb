```python
import pandas as pd 
```

```python
lat_longs = pd.read_csv("CISNC_lat_long.csv")
lat_longs.head()
```

```python
all_schools = pd.read_csv("all_nc_schools.csv")
all_schools.head()
```

```python
CISNC_df = pd.merge(lat_longs, all_schools, how="outer", on=["code", "code"])
CISNC_df.head(10)
```

```python
CISNC_df.describe()
```

```python
CISNC_df.to_csv('merge_all_CISNC.csv')
```
