```python
import pandas as pd 
```

```python
cisnc = pd.read_csv("CISNC_schools.csv")
cisnc.head()
```

```python
df_schools = pd.read_csv("df_schools_geocoded.csv")
df_schools.head()
```

```python
schools_df = pd.merge(cisnc, df_schools, how="outer", on=["name", "name"])
schools_df.head(10)
```

```python
schools_df.describe()
```

```python
schools_df.to_csv('merge_cisnc.csv')
```
